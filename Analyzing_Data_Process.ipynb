# Step 2.

Load your data, typically it is in a CSV format. 

You will use Pandas to load it into a `Dataframe` a special data structures tool.

A `Dataframe` is a visual CSV like representation but has enforced data types.

```
dataframe_name = pd.read_csv(
    'file/location.csv',
    parse_dates=['Date'])

```
To read the date column correctly, we can use the argument parse_dates to specify a list of date columns. Above is the column name 'Date'.

Ways to use `parse_dates`
1. Reading date columns from a CSV file  
  ex) `parse_dates=['date']`
2. Day first format (DD/MM, DD MM or, DD-MM)  
  ex) 
  ```
  pd.read_csv('data/data_1.csv', 
              parse_dates=['date'], 
              dayfirst=True)
  ```
3. Combining multiple columns to a datetime  
  ex)
  ```
  pd.read_csv('data/data_4.csv',
                 parse_dates=[['year', 'month', 'day']])
  ```
4. Customizing a date parser
  ex)
  ```
  from datetime import datetime

  custom_date_parser = lambda x: datetime.strptime(x, "%Y-%d-%m %H:%M:%S")

  df = pd.read_csv('data/data_6.csv',
                 parse_dates=['date'],
                date_parser=custom_date_parser)
  ```
